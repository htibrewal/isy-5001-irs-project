## Pre-processing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
transaction_data_path = '/Users/harsh/Desktop/Intelligent Reasoning Systems/Project/Raw Data/DATA_Cleaned.csv'

data = pd.read_csv(transaction_data_path, encoding='ISO-8859-1')
data.columns = data.columns.str.replace(' ', '_').str.upper()
data.head()
```

```python
data = data.drop(labels=['#', 'DOCCUR', 'SUPPLIER_NAME', 'EXPECTED_DATE', 'FOLLOW_UP_DATE'], axis=1)
```

```python
data.head(10)
```

```python
data[['SUPPLIER_CODE', 'DOC_DATE', 'DELIVERY_DATE', 'ITEM_CODE', 'ITEM_NAME', 'FG_ITEM_DESCRIPTION', 'ITEM_DETAILS', 'MANUFACTURER_PART_NO', 'HSN', 'UOM']].nunique()
```

```python
string_fields = ['SUPPLIER_CODE', 'ITEM_CODE', 'MANUFACTURER_PART_NO', 'HSN', 'UOM']

numerical_fields = ['PO_VALUE', 'ORDERED_QUANTITY', 'PRICE', 'ITEM_VALUE']
```

```python
# convert the string fields from object to string
for field in string_fields:
    data[field] = data[field].astype('str')
    print(field)
    print(data[field].dropna().unique())
    print()
```

```python
# convert the numerical fields from object to float
for field in numerical_fields:
    data[field] = data[field].str.replace(',', '').astype('float64')
```

```python
data.dtypes
```

```python
# data.describe(include=np.number)

np.round(data.describe(include=np.number).transpose(), decimals=2)
```

```python
X_numerical = data[numerical_fields]
X_numerical.corr()
```

```python
Y = data['SUPPLIER_CODE']
```

```python
from sklearn.feature_selection import mutual_info_classif

mi_scores = mutual_info_classif(X_numerical, Y)
mi_scores = pd.Series(mi_scores, name="MI Scores", index=numerical_fields)
mi_scores = mi_scores.sort_values(ascending=False)
mi_scores
```
