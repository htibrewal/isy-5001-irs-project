```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/DATA 1.xlsx - Sheet1.csv')
```

```python
df.head(50)
```

```python
df.columns = df.columns.str.replace(' ', '_').str.upper()
df.shape
```

```python
def find_non_ascii(text):
    return [char for char in text if ord(char) > 127]

# 2. Create a new column that stores the non-ASCII characters for each item name
df['non_ascii_chars'] = df['ITEM_NAME'].apply(find_non_ascii)

# 3. Filter out rows where non-ASCII characters are found
rows_with_non_ascii = df[df['non_ascii_chars'].apply(lambda x: len(x) > 0)]

# 4. Count the frequency of each non-ASCII character across all item names
from collections import Counter
non_ascii_freq = Counter()

for chars in df['non_ascii_chars']:
    non_ascii_freq.update(chars)

# Display rows with non-ASCII characters and the frequency of those characters
print("Rows with non-ASCII characters:\n", rows_with_non_ascii[['ITEM_NAME', 'non_ascii_chars']])
print("\nFrequency of non-ASCII characters:\n", non_ascii_freq)

# If you want to know the number of rows with non-ASCII characters
num_rows_with_non_ascii = len(rows_with_non_ascii)
print(f"\nNumber of rows with non-ASCII characters: {num_rows_with_non_ascii}")
```

```python
specific_signs = ['″']
#, '„', '″'
# Function to check if any of the specific signs are in the item name
def contains_specific_signs(item_name, signs):
    return any(sign in item_name for sign in signs)

# Filter rows where item_name contains any of the specific signs
filtered_df = df[df['ITEM_NAME'].apply(lambda x: contains_specific_signs(x, specific_signs))]

# View the filtered rows
print(filtered_df['ITEM_NAME'])
```

```python
import re

# Dictionary to map non-ASCII characters to text replacements
replacement_dict = {
    '\xa0': '<nbsp>',
    'Â': '<A>',
    '‚': '<low_single_quote>',
    '°': '<deg>',
    'â': '<a>',
    '–': '<dash>',
    '€': '<euro>',
    'Ã': '<A>',
    '®': '<registered>',
    '²': '<superscript_two>',
    '\x90': '<control_char>',
    'Ø': '<O>',
    '¢': '<cent>',
    '’': '<right_single_quote>',
    '±': '<plus_minus>',
    '¸': '<cedilla>',
    '¬': '<not>',
    '„': '<low_double_quote>',
    '¹': '<superscript_one>',
    '™': '<trademark>',
    'Ö': '<O_umlaut>',
    '×': '<multiplication>',
    'ç': '<c_cedilla>',
    '·': '<middle_dot>',
    '\x9d': '<control_char>',
    '˜': '<tilde>',
    '¡': '<inverted_exclamation>',
    'ƒ': '<f>',
    '\u200e': '<left_to_right_mark>',
    'Ÿ': '<Y_diaeresis>',
    '‐': '<hyphen>',
    '…': '<ellipsis>',
    'œ': '<oe>',
    '\x81': '<control_char>',
    'ü': '<u_umlaut>',
    '—': '<em_dash>',
    '½': '<half>',
    '聽': '<unicode_hear>',
    'µ': '<mu>',
    'ä': '<a_umlaut>',
    '″': '<double_prime>',
    '\xad': '<soft_hyphen>',
    '，': '<chinese_comma>',
    '³': '<superscript_three>',
    'º': '<masculine_ordinal>',
    'è': '<e_grave>',
    'é': '<e_acute>',
    '‘': '<left_single_quote>',
    '₹': '<rupee>',
    'Ž': '<Z_caron>',
    '‑': '<non_breaking_hyphen>',
    '•': '<bullet>',
    'ë': '<e_umlaut>',
    '脳': '<unicode_multiply>',
    '◦': '<bullet_point>',
    'ˆ': '<circumflex>',
    '\u2002': '<en_space>'
}

# Function to handle different double quote cases
def replace_quotes(text):
    modified = False  # Track if a modification was made

    # Case 1: Handle numbers in double quotes (INCHES)
    text, num_subs = re.subn(r'“(\d+\.?\d*)”', r'\1<inches>', text)
    if num_subs > 0:
        print(f"Replaced number quotes as inches: {text}")
        modified = True

    # Case 2: Handle right double quotes after a number (INCHES)
    text, num_subs = re.subn(r'(\d+\.?\d*)”', r'\1<inches>', text)
    if num_subs > 0:
        print(f"Replaced standalone inches: {text}")
        modified = True

    # Case 3: Handle quoted text that is not a number (QUOTE)
    text, num_subs = re.subn(r'“([^0-9]+)”', r'QUOTE<\1>', text)
    if num_subs > 0:
        print(f"Replaced quoted text: {text}")
        modified = True

    # Replace any remaining quotes if not already replaced
    if not modified:
        text = text.replace('“', '<left_quote>').replace('”', '<right_quote>')
        #print(f"Replaced unmatched quotes: {text}")

    return text

# Function to replace non-ASCII characters with their text representations
def replace_non_ascii_with_text(text, replacement_dict):
    # Replace non-ASCII characters using the dictionary
    for char, replacement in replacement_dict.items():
        text = text.replace(char, replacement)

    # Apply the special handling for quotes
    text = replace_quotes(text)

    return text

# Apply the function to the 'ITEM_NAME' column
df['CLEANED_ITEM_NAME'] = df['ITEM_NAME'].apply(lambda x: replace_non_ascii_with_text(x, replacement_dict))

# View the original and cleaned item names
print(df[['ITEM_NAME', 'CLEANED_ITEM_NAME']])
```

```python
df['non_ascii_chars'] = df['CLEANED_ITEM_NAME'].apply(find_non_ascii)

non_ascii_freq = Counter()

for chars in df['non_ascii_chars']:
    non_ascii_freq.update(chars)


#print("Rows with non-ASCII characters:\n", rows_with_non_ascii[['CLEANED_ITEM_NAME', 'non_ascii_chars']])
print("\nFrequency of non-ASCII characters:\n", non_ascii_freq)
```

```python
specific_signs = ['<unicode_hear>']
filtered_df = df[df['CLEANED_ITEM_NAME'].apply(lambda x: contains_specific_signs(x, specific_signs))]

# View the filtered rows
print(filtered_df['CLEANED_ITEM_NAME'])
```
