```python
# Loading the data

from google.colab import drive
import pandas as pd
import numpy as np

drive.mount('/content/drive')

# Load the dataset from an Excel file
Sites = pd.read_excel('/content/drive/MyDrive/NUS-ISS AIS Projects/Project 1/Data/DATA 3.xlsx', sheet_name='Data', engine='openpyxl') # Replace 'YourSheetName'

# Display the first few rows of the dataset
print('\n First few rows of the Dataset: \n', Sites.head(), '\n')
```

```python
#Selecting Locations

Site_Loc = ['LOCATION']
Sites = Sites.loc[:, Site_Loc]

print('\n Locations: \n', Sites)
```

```python
#Getting Locations

import pandas as pd
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import time

# Function to get latitude and longitude from an address
def get_coordinates(location_name):
    try:
        geolocator = Nominatim(user_agent="site_locator")
        location = geolocator.geocode(location_name + ", India")
        if location:
            return (location.latitude, location.longitude)
        else:
            return None
    except Exception as e:
        return None

# Function to calculate the distance between two locations
def calculate_distance(site_coords, noida_coords):
    if site_coords:
        distance = geodesic(site_coords, noida_coords).kilometers
        return distance
    else:
        return "Coordinates not found"

#Loading excel file
df = Sites

#Selecting Location
selected_sites = df['LOCATION']

# Get coordinates for Noida, India
noida_coords = get_coordinates('Noida')

# Create a new DataFrame to store the site names and distances from Noida
results = pd.DataFrame(selected_sites)
results['Distance (km)'] = results['LOCATION'].apply(lambda site: calculate_distance(get_coordinates(site), noida_coords))

# Display the result
print(results)

# Optionally, save the result to a new Excel or CSV file
results.to_excel("sites_distances_from_noida.xlsx", index=False)
```
