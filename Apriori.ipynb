```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Create a basket of items for each Purchase Order (PO)
basket = df.groupby(['PO_NUM', 'ITEM_NAME'])['ORDERED_QUANTITY'].sum().unstack().reset_index().fillna(0).set_index('PO_NUM')

# Convert the quantities into a binary format (1 if purchased, 0 otherwise) using map instead of applymap
basket = basket.map(lambda x: 1 if x > 0 else 0)
```

```python
# Apply Apriori algorithm without the 'min_len' argument
frequent_itemsets = apriori(basket, min_support=0.001, use_colnames=True)

# Filter itemsets by length (e.g., only itemsets of length >= 2)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
#filtered_itemsets = frequent_itemsets[frequent_itemsets['length'] >= 2]

# Display the filtered itemsets
print(filtered_itemsets.head(20))

```

```python
print(frequent_itemsets[4800:].head(100))
```

```python
# Check if 'CABLE' is in any frequent itemsets
cable_itemsets = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: 'CABLE' in x)]

# Display the itemsets that contain 'CABLE'
print(cable_itemsets)
```

```python
# Generate association rules with a lower confidence threshold
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.05)

# Display more rules
print(rules.head())
```

```python
# Function to get recommendations based on an input item
def get_recommendations(item):
    # The 'antecedents' and 'consequents' are stored as frozensets, so we need to filter accordingly
    return rules[rules['antecedents'] == frozenset([item])]

# Example: Recommend items when "CABLE" is purchased
recommendations = get_recommendations('CABLE')

# Display the recommendations
print(recommendations)
```

```python
# Sort rules by lift, confidence, or support to evaluate their strength
sorted_rules = rules.sort_values(by='lift', ascending=False)
print(sorted_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(20))
```
